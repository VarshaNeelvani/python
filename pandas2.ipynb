# Time Series Operations

### Time series include the date and time for every record and can fetch the records of dataframe.We can find out the data within a certain range of date and time by using pandas module named Time series.

```python
import numpy as np
import pandas as pd
from datetime import datetime
time = pd.date_range(start = '10/1/2022', end = '11/1/2022', freq = 'Min')
print(time)
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
time = pd.date_range(start = '10/1/2022', end = '11/1/2022', freq = 'Min')
df = pd.DataFrame(time, columns=['DateTime'])
print(df)
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
time = pd.date_range(start = '10/1/2022', end = '11/1/2022', freq = 'Min')
df = pd.DataFrame(time, columns=['DateTime'])
df['DATA'] = np.random.randint(0, 100, size=(len(time)))
print(df)
df.head()
```

# Pandas Pivot

```python
import pandas as pd
from pandas import DataFrame
df1 = DataFrame(
    [
        ['Aryan', 'Singh', 'Full-time','Administration', 2, 20000],
        ['Rohan', 'Agarwal', 'Intern', 'Technical', 3, 5000],
        ['Riya', 'Shah', 'Full-time', 'Administration', 5, 10000],
        ['Yash', 'Bhatia', 'Part-time', 'Technical', 7, 10000],
        ['Siddhanth', 'Khanna', 'Full-time', 'Management', 6, 20000]
    ], columns=['First Name', 'Last Name', 'Type', 'Department', 'YoE', 'Salary']
)
df1
```

```python
#Find the mean value of salary
table = pd.pivot_table(df1, index = "Type", values = "Salary")
table
```

```python
p_table = pd.pivot_table(df1, index="Type",columns="Department",values="Salary")
p_table
```

```python
p_table = pd.pivot_table(df1, index="Type",columns="Department",values="Salary",aggfunc='sum')
p_table
```

```python
table = pd.pivot_table(df1, index="Type")
table
```

# Pandas Melt

```python
import pandas as pd
df = pd.DataFrame([
    ['Amyra', 19, 14, 18, 19],
    ['Neil', 18, 10, 17, 16],
    ['Nitesh', 17, 16, 18, 15]
],
columns=['Name', 'DSP', 'OSA', 'JAVA', 'SE'])

print(df)
df.melt(id_vars=['Name'])
```
