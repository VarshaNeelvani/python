# USER_DATA CSV

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('csv files/user_data.csv')
data.head()
```

```python
x = data.iloc[:,[2,3]].values
y = data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
stx = StandardScaler()
x_train = stx.fit_transform(x_train)
x_test = stx.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

# TITANIC CSV

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv files/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Ticket','Fare','Cabin','Name'], inplace=True, axis=1)
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,1:]
x
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
tr = ColumnTransformer(transformers=[
    ('t2',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked'])
], remainder='passthrough')
```

```python
a = tr.fit_transform(x_train)
a
```

```python
b = tr.fit_transform(x_test)
b
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(a,y_train)
```

```python
y_pred = log.predict(b)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy matrix:", accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
result = pd.DataFrame()
result['Actual label'] = y_test
result['Logistic Prediction'] = y_pred
```

```python
result
```

# IRIS CSV

```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv('csv files/iris.csv')
iris
```

```python
iris.isnull().sum()
```

```python
x = iris.iloc[:,0:3]
x
```

```python
y = iris.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ytr = le.fit_transform(y_train)
yte = le.fit_transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(x_train,ytr)
```

```python
y_pred = log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy score:", accuracy_score(yte,y_pred))
cm = pd.DataFrame(confusion_matrix(yte,y_pred))
cm
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred
```

```python
result
```

```python
from sklearn.metrics import precision_score
precision_score(yte,y_pred,average='weighted')
```

```python
from sklearn.metrics import recall_score
recall_score(yte,y_pred,average='micro')
```

```python
from sklearn.metrics import f1_score
f1_score(yte,y_pred,average='micro')
```
