```python
#Finding the boundary values
ub = df['Sale_amt'].mean() + 3*df['Sale_amt'].std()
lb = df['Sale_amt'].mean() - 3*df['Sale_amt'].std()
print("Highest allowed", ub)
print("Lowest allowed", lb)
```

```python
upper_limit = df['Sale_amt'].mean() + 3*df['Sale_amt'].std()
lower_limit = df['Sale_amt'].mean() - 3*df['Sale_amt'].std()
```

```python
lower_limit
```

```python
upper_limit
```

```python
df['Sale_amt'] = np.where(
    df['Sale_amt'] > upper_limit,
    upper_limit,
    np.where(
        df['Sale_amt'] < lower_limit,
        lower_limit,
        df['Sale_amt']    
    )
)
```
