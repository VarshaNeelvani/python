# HOW TO PUT THE LEGEND OUT OF THE PLOT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame([[10,20],[11,25,],[12,21],[9,18],[8,18]],index=[1,2,3,4,5],columns=['PP','DSP'])
r = df.index
pp = df['PP']
dsp = df['DSP']
plt.title('First Chart', color='r', fontsize=25)
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.ylim([0,40])
plt.xlim([1,7])
plt.xticks(np.arange(1, 7, 1))
plt.yticks(np.arange(0, 40, 10))
plt.plot(r,pp)
plt.plot(r,dsp)
plt.legend(['PP','DSP'],bbox_to_anchor=(1.05, 1.0),loc='upper left')
plt.show()
```
